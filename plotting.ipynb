```python
import matplotlib.pyplot as plt # My prefference
import plotly.express as px     # An option
import altair as alt            # Well ... pretty I guess
import numpy as np
```

```python
# Make some random data
x = np.linspace(0, 10, 1000)
y = np.sin(x)
z = np.sin(x)**2
s = 0.1 * y
```

```python
# Figure canvas and the axes within the canvas
options = dict(
    constrained_layout=True, #Automatically adjusts margins and such
    figsize = (5,3) #Inches
    )
fig, axs = plt.subplots(**options)
axs.set_xlabel("x")
axs.set_ylabel("y")
axs.plot(x, y, label="y")
axs.plot(x, z, label="z")
axs.legend()
```

```python
# Figure canvas and the axes within the canvas
options = dict(
    constrained_layout=True, #Automatically adjusts margins and such
    figsize = (5,3) #Inches
    )
fig, axs = plt.subplots(**options)
axs.set_xlabel("x")
axs.set_ylabel("y")
axs.plot(x, y, label="y")
axs.errorbar(x[::50], y[::50], yerr=np.abs(3*s[::50]), marker ='o', ls='none', mfc='w', elinewidth=1, capsize=3, label="y and errorbar")
axs.legend()
```

```python
# Figure canvas and the axes within the canvas
options = dict(
    constrained_layout=True, #Automatically adjusts margins and such
    figsize = (5,3) #Inches
    )
fig, axs = plt.subplots(**options)
axs.set_xlabel("x")
axs.set_ylabel("y")
axs.plot(x, y, label="y")
axs.fill_between(x, y-s, y+s, color='y', alpha=.3)
axs.legend()
```

```python
# Figure canvas and the axes within the canvas
options = dict(
    nrows=2,
    ncols=2,
    constrained_layout=True, #Automatically adjusts margins and such
    figsize = (6,3) #Inches
    )
fig, axs = plt.subplots(**options)
print(axs)
for i in range(axs.shape[0]):
    for j in range(axs.shape[1]):
        axs[i,j].set_xlabel("x")
        axs[i,j].set_ylabel("y")
        axs[i,j].plot(x, y, label="y")
        axs[i,j].fill_between(x, y-s, y+s, color='y', alpha=.3)
        axs[i,j].legend()
axs
```
